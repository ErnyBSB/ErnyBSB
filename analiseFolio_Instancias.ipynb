# Importar e limpar dados

```python
import pandas as pd

file_path = '/content/drive/MyDrive/DataFromFolio/instancias_fev26.CSV'

# Try reading with semicolon as delimiter and latin-1 encoding
try:
    df = pd.read_csv(file_path, encoding='latin-1', sep=';')
    print("Successfully read with semicolon delimiter and latin-1 encoding.")
    display(df.head())
except Exception as e:
    print(f"Failed to read with semicolon delimiter and latin-1 encoding: {e}")
    # If it still fails, let's peek into the file content again just in case
    print("\nCould not determine delimiter or encoding. Displaying first 5 lines of the file:")
    with open(file_path, 'r', encoding='latin-1', errors='ignore') as f:
        for i, line in enumerate(f):
            print(line.strip())
            if i >= 4:
                break
```

# Exercicio 1 >> Identificar palavras-chave

```python
if 'Instances - Subject headings' in df.columns:
    # Split the 'Instances - Subject headings' column by '--' into new columns
    # n=6 means it will split at most 6 times, resulting in 7 columns (original + 6 splits)
    df_termos = df['Instances - Subject headings'].str.split('--', expand=True, n=6)
    print("Created df_termos with terms split up to 7 columns.")
    display(df_termos)
else:
    print("The column 'Instances - Subject headings' was not found in the DataFrame. Please check the column name.")
    print("Available columns:", df.columns.tolist())
```

```python
# Stack all columns of df_termos into a single Series, dropping NaN values
termos_series = df_termos.stack().reset_index(drop=True)

# Count the occurrences of each unique term
df_termos_ordenado = termos_series.value_counts().reset_index()

# Rename the columns for clarity
df_termos_ordenado.columns = ['termos', 'contagem']

# Display the sorted DataFrame
print("Created df_termos_ordenado with terms and their counts, sorted descending by count.")
display(df_termos_ordenado)
```

```python
import plotly.express as px

# Get the top 100 terms from df_termos_ordenado
df_top_100_termos = df_termos_ordenado.head(10)

# Create the bubble chart
fig = px.scatter(
    df_top_100_termos,
    x=range(len(df_top_100_termos)), # Use index for x-axis to spread points, or a dummy variable
    y='contagem',
    size='contagem',
    color='termos',
    hover_name='termos',
    text='termos',
    size_max=60,
    title='Top 100 Termos Mais Ocorrentes (Gráfico de Bolhas)',
    labels={'x': 'Posição no Ranking', 'contagem': 'Contagem de Ocorrências'}
)

fig.update_traces(textposition='top center')
fig.update_layout(
    xaxis_showgrid=False, # Hide x-axis grid lines
    yaxis_showgrid=True, # Show y-axis grid lines
    xaxis_showticklabels=False, # Hide x-axis tick labels for a cleaner look
    showlegend=False # Do not use legend names outside the bubbles
)

fig.show()
```
